```python
# put all alternate titles in a dictionary
# get all the keys referencing alternate titles
# function will operate on one movie dictionary at a time.
# iterate through all the alt title keys
#   if the key is a member of the movie dictionary
#     add key, value pair to alt_titles dictionary
#     remove key, value pair from movie dictionary
# when finished the list of keys, check how many alt titles were found
# if any alt title were found add the new alt_titles dictionary to the movie dictionary
# return the revised movie dictionary
def clean_movie(movie):
    movie = dict(movie) #create a non-destructive copy
    alt_titles = {}
    for key in ['Also known as','Arabic','Cantonese','Chinese','French',
                'Hangul','Hebrew','Hepburn','Japanese','Literally',
                'Mandarin','McCune–Reischauer','Original title','Polish',
                'Revised Romanization','Romanized','Russian',
                'Simplified','Traditional','Yiddish']:
        if key in movie:
            alt_titles[key] = movie[key]
            movie.pop(key)
    if len(alt_titles) > 0:
        movie['alt_titles'] = alt_titles
    return movie   
```

```python
# modify patterns to get more matches
# fix number ranges by keeping just the upper end of the range
# ends with zero to one period
box_office = box_office.str.replace('\$.*[-—–](?![a-z])', '$', regex=True)

# then extract only the parts of the strings that match the patterns
form_one = '\$\s*\d+\.?\d*\s*[mb]illi?on'
form_two = '\$\s*\d{1,3}(?:[,\.]\d{3})+(?!\s[mb]illion)'
# combine into one capture group using pipe
box_office.str.extract(f'({form_one}|{form_two})')
```

```python
# change strings into numerics using a parsing function
# use re.match(pattern, string) to determine whether string matches the pattern)
def parse_dollars(s):
    # if s is not a string, return NaN
    if type(s) != str:
        return np.nan
   
    # split the millions and billions patterns
    # form1: if input is of the form $###.# million
    # re.match returns the first match it finds in the string
    if re.match(r'\$\s*\d+\.?\d*\s*milli?on', s, flags=re.IGNORECASE):
        # remove dollar sign and " million"
        s = re.sub('\$|\s|[a-zA-Z]','',s)
        # convert to float and multiply by a million
        value=float(s) * 10**6
        # return value
        return value
   
    # form1: if input is of the form $###.# billion
    elif re.match(r'\$\s*\d+\.?\d*\s*billi?on', s, flags=re.IGNORECASE):
        # remove dollar sign and " billion"
        s = re.sub('\$|\s|[a-zA-Z]','', s)
        # convert to float and multiply by a billion
        value = float(s) * 10**9
        # return value
        return value
    
    # form2: if input is of the form $###,###,###
    elif re.match(r'\$\s*\d{1,3}(?:[,\.]\d{3})+(?!\s[mb]illion)', s, flags=re.IGNORECASE):
        # remove dollar sign and commas
        s = re.sub('\$|,','', s)
        # convert to float
        value = float(s)
        # return value
        return value
    
    # otherwise, return NaN
    else:
        return np.nan
    
wiki_movies_df['box_office'] = box_office.str.extract(f'({form_one}|{form_two})', 
flags=re.IGNORECASE)[0].apply(parse_dollars)
```

```python
# extract the values, digits only
# pattern 1: (one or more digits), optional spaces, h, optional ours, optional spaces, (optional digits)
# pattern 2: (at least one digit), optional spaces, required m
# three capture groups
running_time_extract = running_time.str.extract(r'(\d+)\s*ho?u?r?s?\s*(\d*)|(\d+)\s*m')
```

```python
# convert strings to numeric 
# use the .to_numeric() method with error argument set to 'coerce', which turns empty strings into (NaN)
# then use fillna() to change NaNs to zeros

# method apply(function,iterator), lambda function returns an iterator
# form: running_time_extract = running_time_extract.apply(lambda).fillna(0)

# syntax: lambda arguments: expression
# arguments: col
# expression: (the function) pd.to_numeric(col,error='coerce') ==> returns an iterator
# lambda col: pd.to_numeric(col, errors='coerce'))

running_time_extract = running_time_extract.apply(lambda col: pd.to_numeric(col, errors='coerce')).fillna(0)
```

```python
# where col2 is 0, convert col0 hours to minutes and add col1 minutes to get total minutes
# another lambda function:
# syntax: lambda arguments: expression
# arguments: row
# expression: (the function) calculate the minutes to return

#(lambda row: row[0]*60 + row[1] if row[2]==0 else row[2], axis=1) # axis=1 means columns: so apply function to each row.

# use the apply method to apply lambda to each col/series

wiki_movies_df['running_time'] = running_time_extract.apply(lambda row: row[0]*60 + row[1] if row[2]==0 else row[2], axis=1)
```
